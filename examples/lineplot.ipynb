```python
from nnterp import StandardizedTransformer
from nnsightful import logit_lens

model = StandardizedTransformer("gpt2")
data = logit_lens.logit_lens(model, "The capital of France is", top_k=5)
data.display()
```

### Logit lens with custom layer and position ranges

You can use `IndexSpec` to select specific layers and positions. Supported formats:
- `int` — single index (negative indices supported)
- `(start, end)` — range `[start, end)`
- `[int, ...]` — explicit list
- `[(start, end), ...]` — union of ranges

```python
# Only the first 4 and last 2 layers, only the last 3 token positions
data_sparse = logit_lens.logit_lens(
    model,
    "The capital of France is",
    layers=[(0, 4), (-2, -1)],  # union of two ranges
    positions=[-3, -2, -1],     # last 3 positions
    top_k=5,
    top_p=0.9,                  # nucleus selection (union with top_k)
)
print(f"Layers: {data_sparse.layers}")
print(f"Positions: {data_sparse.positions}")
print(f"Tracked tokens at last position: {list(data_sparse.tracked[-1].keys())}")
data_sparse.display()
```
